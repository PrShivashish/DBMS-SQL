```python
#creating a new table called 'class'
CREATE TABLE class (
    id INT,
    name VARCHAR(50),
    age INT,
    course VARCHAR(50),
    marks INT
);
insert into class values( 1, "Manish" , 18, "SQL", 78); #inserting values just 1
insert into class values( 2, "Shivashish" , 20, "SQL", 98), (3, "Srijan", 20, "SQL", 95), (4, "Ayush", 20, "SQL", 92); #inserting multiple values
SELECT * FROM class;
SELECT name, marks FROM class;
SELECT * FROM class ORDER BY marks DESC; #descend
SELECT * FROM class ORDER BY marks ASC; #ascend
SELECT * FROM class WHERE marks > 90;
SELECT AVG(marks) FROM class;
SELECT MIN(marks) FROM class;
SELECT COUNT(*) FROM class;
#OrderBy -> Sorting
#Sum(marks) -> Sum of marks
#AVG(makrs) -> Avg marks
SELECT marks FROM class WHERE id = 2 AND marks = 98;
DELETE FROM class WHERE id=3;
SELECT * FROM class ORDER BY marks ASC LIMIT 2;#printing the top
UPDATE class SET marks=33 where id=1;
```

***DELETE:*** Only one deleted data deleted

***TRUNCATE:*** Only data deleted but we have table exists, TRUNCATE TABLE class;

***DROP:*** Whole table deleted from database, DROP TABLE class;

```python
select * from class;
select course, avg(marks) from class group by course;
update class set course="python" where marks = 98;
update class set course="java" where marks = 92;
update class set course="c++" where marks = 75;
update class set course="DSA" where marks = 50;
select course, avg(marks) from class group by course;
select avg(marks), max(marks) from class group by age;
select avg(marks), max(marks) from class group by age,course;
```

```python
#Give me average age and minimum marks by name and id
select name, id, avg(age), min(marks) from class group by name, id;
```
